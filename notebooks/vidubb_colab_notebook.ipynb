# Clone ViDubb Repository

```python
import os
import shutil
import subprocess

# Set the path where you want to clone the repository
repo_dir = "ViDubb"

# Check if the directory exists
if os.path.exists(repo_dir):
    print("Directory exists. Deleting it...")
    shutil.rmtree(repo_dir)  # Deletes the directory and its contents

# Clone the repository
repo_url = "https://github.com/medahmedkrichen/ViDubb.git"
print(f"Cloning repository from {repo_url}...")
subprocess.run(["git", "clone", repo_url])
```

# Configuration

```python
# Update the HF_TOKEN value (required)
env_file_path = 'ViDubb/.env'
with open(env_file_path, 'r') as file:
    lines = file.readlines()


new_token = ""
for i, line in enumerate(lines):
    if line.startswith('HF_TOKEN'):
        lines[i] = f'HF_TOKEN={new_token}\n'

# Write the updated lines back to the .env file
with open(env_file_path, 'w') as file:
    file.writelines(lines)



# Update the Groq_TOKEN value (optional)
env_file_path = 'ViDubb/.env'

# Read the contents of the .env file
with open(env_file_path, 'r') as file:
    lines = file.readlines()

new_token = ""
for i, line in enumerate(lines):
    if line.startswith('Groq_TOKEN '):
        lines[i] = f'Groq_TOKEN ={new_token}\n'

# Write the updated lines back to the .env file
with open(env_file_path, 'w') as file:
    file.writelines(lines)
```
